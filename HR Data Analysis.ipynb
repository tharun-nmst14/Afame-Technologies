```python
# Load the HR dataset
file_path = "HR Data.csv"
hr_data = pd.read_csv(file_path)
```

```python
# Display basic info about the dataset
print("Dataset Info:")
print(hr_data.info())
print("\nFirst few rows:")
print(hr_data.head())
```

```python
# Remove duplicate rows
hr_data.drop_duplicates(inplace=True)
hr_data.head()  
```

```python
# Rename columns for better readability (example)
hr_data.rename(columns={
    "Emp ID": "Employee_ID",
    "Dept": "Department",
    "Salary": "Annual_Salary"
}, inplace=True)
hr_data.head()  
```

```python
# Handle missing values - Fill or Drop
hr_data.dropna(inplace=True)  # Drop rows where Employee_ID is missing
hr_data.fillna(method='ffill', inplace=True)  # Forward fill other missing values
```

```python
# Convert categorical columns to consistent format
if "Department" in hr_data.columns:
    hr_data["Department"] = hr_data["Department"].str.strip().str.title()
hr_data.head()    
    
```
